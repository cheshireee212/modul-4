```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# URL dataset Mushroom dari UCI Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data"

# Memuat dataset menjadi objek DataFrame oleh Pandas
nama_kolom = [
    "cap-shape", "cap-surface", "cap-color", "bruises", "odor",
    "gill-attachment", "gill-spacing", "gill-size", "gill-color",
    "stalk-shape", "stalk-root", "stalk-surface-above-ring",
    "stalk-surface-below-ring", "stalk-color-above-ring",
    "stalk-color-below-ring", "veil-type", "veil-color",
    "ring-number", "ring-type", "spore-print-color", "population",
    "habitat", "class"
]

# Memuat dataset dan menetapkan nama kolom
data = pd.read_csv(url, header=None, names=nama_kolom)
    
```

```python
def diagram_lingkaran():
    bruises_counts = data['bruises'].value_counts() # Membuat diagram lingkaran 
    plt.figure(figsize=(8, 6)) , bruises_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999']) 
    plt.title('Distribusi Bruises pada Jamur') 
    plt.ylabel('') 
    # Menghilangkan label y 
    plt.show()
diagram_lingkaran()
```

```python

def melihat_data():
    # Melihat beberapa baris pertama dari dataset
    print(data.iloc[:, :2].head(6)) #nampilinn data 2 kolom aja
    masukan=input("lihat lebih banyak ? y  : ")
    if masukan.lower() == "y" or masukan.upper() == "y":
        lihat = int(input("masukan data yang ingin dilihat : "))
        print(f"{data.head(lihat)}")
        melihat_data()
    else:
        print("ok")
```

```python
# Menghitung frekuensi nilai dalam kolom 'odor' 
odor_counts = data['odor'].value_counts() # Membuat diagram batang 
plt.figure(figsize=(10, 6)) 
odor_counts.plot(kind='bar', color='skyblue') 
plt.title('Distribusi Jenis Bau pada Jamur') 
plt.xlabel('Jenis Bau') 
plt.ylabel('Jumlah Data') 
plt.xticks(rotation=45) 
plt.show()
```
